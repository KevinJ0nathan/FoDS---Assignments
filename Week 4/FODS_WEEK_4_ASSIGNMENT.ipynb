# 12. Practice imputing missing values using the SimpleImputer Class in sklearn

```python
from sklearn.impute import SimpleImputer
```

```python
# Initialize the SimpleImputer
# For numerical data, replaces the null value with the mean
imputer_mean = SimpleImputer(strategy='mean')

df_test_dummy_si = df_movies.head(10).copy()
df_test_dummy_si.loc[1,'movie_budget'] = np.nan

df_test_dummy_si['movie_budget'].head()
```

```python
df_test_dummy_si['movie_budget'] = imputer_mean.fit_transform(df_test_dummy_si[['movie_budget']])
df_test_dummy_si['movie_budget'].head()
```

```python
df_test_dummy_si.loc[1,'original_language'] = np.nan
df_test_dummy_si.loc[2,'original_language'] = "fr"
df_test_dummy_si.loc[3,'original_language'] = "cn"
df_test_dummy_si['original_language'].head()
```

```python
 # For categorical data, replaces the null value with the most common value/string
imputer_freq = SimpleImputer(strategy='most_frequent')
df_test_dummy_si['original_language'] = imputer_freq.fit_transform(df_test_dummy_si[['original_language']]).flatten()
df_test_dummy_si['original_language'].head()
```

## 13. Cleaning Text Data using Regex

```python
import re
import json

df_movies = pd.read_csv('/content/tmdb_5000_movies.csv')

df_regex_dummy = df_movies.copy()

#Define a regex pattern to extract the names
pattern = r'"name":\s*"((?:[^"\\]|\\.)*)"'

# Function to extract language names using JSON parsing
def extract_language_names(json_str):
  # Check if the input is a string
    if isinstance(json_str, str):
        try:
            # Parse the JSON string
            languages = json.loads(json_str)
            # Extract names from the parsed JSON
            return [lang['name'] for lang in languages]
        except json.JSONDecodeError:
           # Return empty if JSON decoding fails
            return []
    #Return an empty list for non-string inputs
    return []

# Apply the function to the 'spoken_languages' column
df_regex_dummy['language_names'] = df_regex_dummy['spoken_languages'].apply(extract_language_names)

df_regex_dummy['language_names'] = df_regex_dummy['language_names'].apply(lambda x: ', '.join(x) if x else 'No languages found')

# Display the updated DataFrame
print(df_movies[['spoken_languages', 'language_names']])
```
